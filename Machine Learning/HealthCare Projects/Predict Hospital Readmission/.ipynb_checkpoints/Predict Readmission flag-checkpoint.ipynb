Weight is missing in over 98% records. Owing to the poor interpretability of missing values and little predictive generalizability to other patients, best thing is to just drop it.
Payer code and Medical Specialty of treating physician also have 40–50% missing values. We decided to drop these, but there are other ways too to deal with such missing values.

```python
cols_to_drop = ['weight','payer_code','medical_specialty']
df_process.drop(cols_to_drop, inplace=True, axis=1)
```

* Primary (diag_1), Secondary (diag_2) and Additional (diag_3) diagnoses were have very few missing values. Technically, if all three are missing, that’s bad data. So we only drop those records where all three diagnoses are missing.
* Gender has only 3 missing or invalid values so we decided to drop these records.
* Also, one more cleaning step that depends on understanding the data and some common sense: since we are trying to predict readmissions, those patients who died during this hospital admission, have zero probability of readmission. So we should remove those records (discharge_disposition = 11).

```python
idx_to_drop= set(df[(df_process['diag_1']=='?')  & (df_process['diag_2']=='?') & (df_process['diag_3']=='?')].index)
idx_to_drop= idx_to_drop.union(df[df['gender']=='Unknown/Invalid'].index)
idx_to_drop= idx_to_drop.union(df[df['discharge_disposition_id']==11].index)
```

```python
idx_to_drop
```

```python
new_idx =list(set(df_process.index) - set(idx_to_drop))
```

```python
df_process= df_process.iloc[new_idx]
```

* We also noticed that for two variables (drugs named citoglipton and examide), all records have the same value. So essentially these cannot provide any interpretive or discriminatory information for predicting readmission, and we dropped these columns as well. Technically, this isn’t a missing value problem but rather a missing information problem.

```python
df_process.drop(['citoglipton','examide'], axis=1, inplace=True)
```

Creating and/or Recoding New Features

```python
df_process['service_utilization'] = df_process['number_inpatient']+df_process['number_outpatient']+df_process['number_emergency']
```

```python
keys = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'glipizide', 'glyburide', 'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'insulin', 'glyburide-metformin', 'tolazamide', 'metformin-pioglitazone','metformin-rosiglitazone', 'glimepiride-pioglitazone', 'glipizide-metformin', 'troglitazone', 'tolbutamide', 'acetohexamide']
```

```python
for c in keys:
    col_name = str(c) + 'temp'
    df_process[col_name]=df_process[c].apply(lambda x: 0 if (x=='No' or x=='Steady') else 1 )
```

```python
df_process['numchange']=0
for col in keys:
    colname = str(col) + 'temp'
    df_process['numchange'] = df_process['numchange'] + df_process[colname]
    del df_process[colname]
```

```python
df_process['numchange'].value_counts()
```

```python
df_process['nummed'] = 0
for col in keys:
    df_process['nummed'] = df_process['nummed'] + df_process[col]
```

```python
for index, row in df_process.iterrows():
    row['diag_1']
```

```python
import numpy as np
```

```python
# create a duplicate of the diagnosis column
df_process['level1_diag1'] = df_process['diag_1']
# disease codes starting with V or E are in “other” category; so recode them to 0
df_process.loc[df_process['diag_1'].str.contains('V'), ['level1_diag1']] = 0
df_process.loc[df_process['diag_1'].str.contains('E'), ['level1_diag1']] = 0
# also replace the unknown values with -1
df_process['level1_diag1'] = df_process['level1_diag1'].replace('?', -1)
df_process['level1_diag1']=df_process['level1_diag1'].astype(float)
# iterate and recode disease codes between certain ranges to certain categories
for index, row in df_process.iterrows():
    #print ((row['level1_diag1'].astype('int')))
    if (row['level1_diag1'] >= 390 and row['level1_diag1'] < 460) or (np.floor(row['level1_diag1']) == 785):
        df_process.loc[index, 'level1_diag1'] = 1
    elif (row['level1_diag1'] >= 460 and row['level1_diag1'] < 520) or (np.floor(row['level1_diag1']) == 786):
        df_process.loc[index, 'level1_diag1'] = 2
    elif (row['level1_diag1'] >= 520 and row['level1_diag1'] < 580) or (np.floor(row['level1_diag1']) == 787):
        df_process.loc[index, 'level1_diag1'] = 3
    elif (np.floor(row['level1_diag1']) == 250):
        df_process.loc[index, 'level1_diag1'] = 4
    elif (row['level1_diag1'] >= 800 and row['level1_diag1'] < 1000):
        df_process.loc[index, 'level1_diag1'] = 5
    elif (row['level1_diag1'] >= 710 and row['level1_diag1'] < 740):
        df_process.loc[index, 'level1_diag1'] = 6
    elif (row['level1_diag1'] >= 580 and row['level1_diag1'] < 630) or (np.floor(row['level1_diag1']) == 788):
        df_process.loc[index, 'level1_diag1'] = 7
    elif (row['level1_diag1'] >= 140 and row['level1_diag1'] < 240):
        df_process.loc[index, 'level1_diag1'] = 8
    else:
        df_process.loc[index, 'level1_diag1'] = 0
# convert this variable to float type to enable computations later
df_process['level1_diag1'] = df_process['level1_diag1'].astype(float)
```

```python
df_process['admission_type_id'] = df_process['admission_type_id'].replace(2,1)
df_process['admission_type_id'] = df_process['admission_type_id'].replace(7,1)
df_process['admission_type_id'] = df_process['admission_type_id'].replace(6,5)
df_process['admission_type_id'] = df_process['admission_type_id'].replace(8,5)
```

```python
df_process['change'] = df_process['change'].replace('Ch', 1)
df_process['change'] = df_process['change'].replace('No', 0)
df_process['gender'] = df_process['gender'].replace('Male', 1)
df_process['gender'] = df_process['gender'].replace('Female', 0)
df_process['diabetesMed'] = df_process['diabetesMed'].replace('Yes', 1)
df_process['diabetesMed'] = df_process['diabetesMed'].replace('No', 0)
# keys is the same as before
for col in keys:
    df_process[col] = df_process[col].replace('No', 0)
    df_process[col] = df_process[col].replace('Steady', 1)
    df_process[col] = df_process[col].replace('Up', 1)
    df_process[col] = df_process[col].replace('Down', 1)
```

```python
df_process['A1Cresult'] = df_process['A1Cresult'].replace('>7', 1)
df_process['A1Cresult'] = df_process['A1Cresult'].replace('>8', 1)
df_process['A1Cresult'] = df_process['A1Cresult'].replace('Norm', 0)
df_process['A1Cresult'] = df_process['A1Cresult'].replace('None', -99)
df_process['max_glu_serum'] = df_process['max_glu_serum'].replace('>200', 1)
df_process['max_glu_serum'] = df_process['max_glu_serum'].replace('>300', 1)
df_process['max_glu_serum'] = df_process['max_glu_serum'].replace('Norm', 0)
df_process['max_glu_serum'] = df_process['max_glu_serum'].replace('None', -99)
```

```python
df_process['readmitted'] = df_process['readmitted'].replace('>30', 0)
df_process['readmitted'] = df_process['readmitted'].replace('<30', 1)
df_process['readmitted'] = df_process['readmitted'].replace('NO', 0)
```

```python
# log transformation
```

```python
numerics = [col for col in df2.columns if df2[col].dtype!='object' ]
numerics
```

```python
df2.shape[0]
```

```python
log_numerics = []
for col in numerics:
    if df2[df2[col]==0].shape[0]/df2.shape[0] <0.02:
        df2[col+'logp']=np.log(df2[col])
        log_numerics.append(col+'logp')
    else:
        df2[col+'log1p']=np.log1p(df2[col])
        log_numerics.append(col+'log1p')
log_numerics
```

```python
def standardize(raw_data):
    return ((raw_data - np.mean(raw_data, axis = 0)) / np.std(raw_data, axis = 0))
# numerics is a list of all numeric features
df2[log_numerics] = standardize(df2[log_numerics])
```
