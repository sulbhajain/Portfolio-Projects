```python
import pandas as pd
import numpy as np
```

```python
file_health = pd.read_csv("Data/Current BCHI Platform Dataset - BCHI, Phase I & II.csv")
file_health.head()
```

```python
file_health["Indicator Category"].unique()
```

```python
cancer_file = file_health[file_health["Indicator Category"]=='Cancer']
```

```python
cancer_file.head()
```

```python
cancer_2010 = cancer_file[cancer_file['Year']==2010 ]
cancer_2010.head()
```

```python
cancer_2010_both = cancer_2010[ cancer_2010['Sex']=='Both']
cancer_2010_both
```

```python
cancer_2010_both_boston = cancer_2010_both[ cancer_2010_both['Place']=='Boston, MA']
cancer_2010_both_boston
```

```python
%matplotlib inline

cols = ['Race/Ethnicity', 'Value']
df_sel = cancer_2010_both_boston[cols]
df_sel['Value'] = pd.to_numeric(df_sel['Value'])
df_sel

#scatter(x= 'Race/Ethnicity', y= 'Value')
```

```python
# which Race/Ethnicy is more susceptable to Cancer in 2010
df_sel.plot.bar(x="Race/Ethnicity", y= 'Value')
```
