```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data import
file_health = pd.read_csv("Data/Current BCHI Platform Dataset - BCHI, Phase I & II.csv")
file_health.head()
```

```python
file_health["Indicator Category"].unique()
```

## EDA 

```python
# extract a dataframe for 'cancer' data as exploration will be done for this
cancer_file = file_health[file_health["Indicator Category"]=='Cancer']
```

```python
cancer_file.head()
```

```python
## Question: what is the distribution for 'Cancer' patients as per ethnicity ?
```

```python
# filter the data for the year '2010' 
cancer_2010 = cancer_file[cancer_file['Year']==2010 ]
cancer_2010.head()
```

```python
# consider 'Both' males and females to address the question

cancer_2010_both = cancer_2010[ cancer_2010['Sex']=='Both']
cancer_2010_both.head()
```

```python
# take a subset for ''Boston, MA'
cancer_2010_both_boston = cancer_2010_both[ cancer_2010_both['Place']=='Boston, MA']
cancer_2010_both_boston.head()
```

```python
# plot the distribution

%matplotlib inline

cols = ['Race/Ethnicity', 'Value']
df_sel = cancer_2010_both_boston[cols]
df_sel['Value_num'] = pd.to_numeric(df_sel['Value'])
df_sel.head()

```

```python
# which Race/Ethnicy is more susceptable to Cancer in 2010 ?
df_sel.plot.bar(x="Race/Ethnicity", y= 'Value_num')
```
