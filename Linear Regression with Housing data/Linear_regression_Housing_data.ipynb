## Train a linear regression model

```python
USA_housing =pd.read_csv('USA_Housing.csv')
USA_housing.columns

```

```python
X = USA_housing[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X
```

```python
y = USA_housing['Price']
```

## Train and Test Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
#X_train.loc[ [X_train['Address']=='304 Obrien Village Suite 750\nWest Maureen, AS 63430'], 6]
lm.fit(X_train,y_train)
```

## Model evaluation for coefficients and intercept

```python
# print the intercept
print(lm.intercept_)
```

```python
print (lm.coef_)
```

### Make the coeffecient data frame

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

#### Predictions from our data Model

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

#### Residual histogram

```python
sns.distplot((y_test - predictions ), bins =50)
```

### Regression evaluation matracis:
#### MAE - Mean absolute error
#### MSE - Mean Squared error 
#### RMSE - Root mean squared error

##### Above losses needs to be minimized

```python
from sklearn import metrics
```

```python
print ("MAE", metrics.mean_absolute_error(y_test,predictions))
```

```python
print ("MSE", metrics.mean_squared_error(y_test,predictions))
```

```python
print ("RMSE", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
