### Training and Test Data. Use all numerical values to train against Yearly Time spent

```python
y = customers['Yearly Amount Spent']
```

```python
X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm  = LinearRegression()
```

###  Train/fit on the training data

```python
lm.fit(X_train,y_train)
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
```

```python
df_coeff = pd.DataFrame(lm.coef_, X.columns, columns = ['Coefficients'])
```

```python
df_coeff
```

### Predicting Test Data

```python
predictions = lm.predict(X_test)
```

### create the scatter plot to see variation between predictions and actual Y value

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

### Evaluate the Model

```python
from sklearn import metrics
```

```python
print(" Mean Absolute error:", metrics.mean_absolute_error(y_test, predictions))
```

```python
print(" Mean Squared error:", metrics.mean_squared_error(y_test, predictions))
```

```python
print(" Root Mean Squared error:", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### Residuals

```python
sns.distplot((y_test-predictions),bins=50);
```
