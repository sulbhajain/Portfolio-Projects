```python
import pandas as pd
```

```python
data = pd.read_csv("data/supermarkets.csv")
data
```

```python
from geopy.geocoders import Nominatim
dir(geopy)
```

```python
nominatm = Nominatim()
```

```python
data
```

```python
data['Complete-Adress'] = data['Address'] +' , ' + data['City'] + ' ,' + data['State']
data
```

```python
data['Coordinates'] = data['Complete-Adress'].apply(nominatm.geocode)
data
```

```python
data['Coordinates'][0]
```

```python
data['Latitude'] = data['Coordinates'].apply(lambda x: x.latitude if x != None else None)
data
```

```python
data['Longitude'] = data['Coordinates'].apply(lambda x: x.longitude if x != None else None)
data
```
